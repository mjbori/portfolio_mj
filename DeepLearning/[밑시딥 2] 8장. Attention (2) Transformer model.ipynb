## * RNN(LSTM)

- 사용처: 언어모델, 문장생성, seq2seq, Attention을 갖춘 seq2seq

- 장점: RNN을 통해 가변 길이 시계열 데이터를 잘 처리하여 많은 경우 좋은 결과를 얻음

- 단점: 병렬처리가 불가능

### => RNN을 없애는 연구 or 병렬처리 가능한 RNN 연구

- Transformer 모델

- RNN 대신 합성곱을 이용한 seq2seq 구성 (병렬화 가능)

## * 셀프 어텐션(self-attention)

- 자신에 대한 주목

- 하나의 시계열 데이터를 대상으로 한 어텐션

- 하나의 시계열 데이터 내에서 각 원소가 다른 원소들과 어떻게 관련되는지 살펴보기


## * 어텐션과 셀프어텐션의 비교



- `어텐션`에서 '번역' 과 같이 **2개의 시계열 데이터 사이의 대응관계**를 구함

- `Time Attention` 계층에는 서로 다른 두 시계열 데이터가 입력됨

- `Self-Attention`은 두 입력선이 모두 하나의 시계열 데이터로부터 나옴


## * 트랜스포머 구성



- 트랜스포머에서는 RNN대신 Attention을 사용

- Encoder와 Decoder에서 모두 Self-Attention을 사용

- FeedForward 계층은 피드포워드 신경망(시간 방향으로 독립적으로 처리하는 신경망)을 나타냄 (은닉층이 1개이고 활성화 함수로 ReLU를 이용한 완전연결계층 신경망을 이용)

- Nx(보라색 영역)은 보라색 영역의 계층들을 N겹 쌓았다는 뜻

* 트랜스포머를 이용하면 계산량을 줄이고 GPU를 이용한 병렬 계산의 혜택도 더 많이 누릴 수 있음 -> GNMT보다 학습 시간 대폭 줄임, 번역 품질도 상당 폭 올림
