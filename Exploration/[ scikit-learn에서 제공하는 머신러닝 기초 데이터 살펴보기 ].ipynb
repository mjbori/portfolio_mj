```python
import numpy as np      # 수학적 계산 편하게 사용가능
import pandas as pd     # 데이터를 데이터프레임 형태로 사용가능
import seaborn as sns                # 시각화
import matplotlib.pyplot as plt      # 시각화
from sklearn import datasets         #  scikit-learn에서 제공하는 데이터 사용하기 위해
```

# 1. Boston_house_datasets

- 13가지 피쳐를 이용하여 보스턴 집값을 예측할 때 사용하는 데이터셋

- https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_boston.html#sklearn.datasets.load_boston

```python
'''data 불러오기'''

raw_boston = datasets.load_boston()
X_boston = pd.DataFrame(raw_boston.data)               # .data -> x
y_boston = pd.DataFrame(raw_boston.target)             # .target -> y
df_boston = pd.concat([X_boston, y_boston], axis=1)    # pd.concat([A,B],axis) -> A와 B를 가로축(axis=1) or 세로축(axis=0) 으로 합치기
df_boston.head()
```

```python
'''열이름 재설정'''
features_boston = raw_boston.feature_names          # .feature_name : 피쳐 이름 목록
print(features_boston)
# targetname_boston = raw_boston.target_name -> target_name은 없음
# print(targetname_boston)
```

    ----- x ------------------------------------------------------------------------

    ZN       proportion of residential land zoned for lots over 25,000 sq.ft.

    LSTAT    % lower status of the population 
    ---------------------------------------------------------------------------------

    ----- y -------------------------------------------------------------------------
    target (MEDV)     Median value of owner-occupied homes in 1000's dollar 
    ---------------------------------------------------------------------------------
    



```python
col_names_boston = np.append(features_boston,['MEDV(target)'])
df_boston.columns = col_names_boston                        # .columns : 컬럼 이름 바꾸기
df_boston.head()

# 컬럼 이름, 인덱스 이름 바꾸기: https://rfriend.tistory.com/468
```

# 2. Iris_datasets

- 아이리스의 피처 특성에 따라 꽃의 종류를 분류할 때 쓰이는 데이터셋

- https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html#sklearn.datasets.load_iris

```python
raw_iris = datasets.load_iris()
X_iris = pd.DataFrame(raw_iris.data)
y_iris = pd.DataFrame(raw_iris.target)
df_iris = pd.concat([X_iris, y_iris], axis=1)
df_iris.head()
```

```python
print(raw_iris.DESCR) # .DESCR: Description 데이터셋의 정보를 확인
```

```python
features_iris = raw_iris.feature_names
print("* iris_features name: ",features_iris)
target_iris = raw_iris.target_names
print("* iris_target names: ",target_iris)
```

```python
col_names_iris = np.append(features_iris,['target'])
df_iris.columns = col_names_iris
df_iris.head()
```
